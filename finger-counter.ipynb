```python
def get_sign_right(results,k):
    if results.multi_hand_landmarks[k].landmark[8].y<results.multi_hand_landmarks[k].landmark[7].y:
        if results.multi_hand_landmarks[k].landmark[12].y<results.multi_hand_landmarks[k].landmark[11].y:
            if results.multi_hand_landmarks[k].landmark[16].y<results.multi_hand_landmarks[k].landmark[15].y:
                if results.multi_hand_landmarks[k].landmark[20].y<results.multi_hand_landmarks[k].landmark[19].y:
                    if results.multi_hand_landmarks[k].landmark[4].x<results.multi_hand_landmarks[k].landmark[3].x:
                        return("five")
                    else:
                        return("four")
                else:
                    return("three")
            else:
                return("two")
        else:
            return("one")
    else :
        return("zero")
    
def get_sign_left(results,k):
    if results.multi_hand_landmarks[k].landmark[8].y<results.multi_hand_landmarks[k].landmark[7].y:
        if results.multi_hand_landmarks[k].landmark[12].y<results.multi_hand_landmarks[k].landmark[11].y:
            if results.multi_hand_landmarks[k].landmark[16].y<results.multi_hand_landmarks[k].landmark[15].y:
                if results.multi_hand_landmarks[k].landmark[20].y<results.multi_hand_landmarks[k].landmark[19].y:
                    if results.multi_hand_landmarks[k].landmark[4].x>results.multi_hand_landmarks[k].landmark[3].x:
                        return("five")
                    else:
                        return("four")
                else:
                    return("three")
            else:
                return("two")
        else:
            return("one")
    else :
        return("zero")
```

```python
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

chg,chr="none","none" #each string is representing the number on one hand

cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    model_complexity=0,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      break

    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    flipped=cv2.flip(image, 1)
    results = hands.process(flipped)


    flipped.flags.writeable = True
    image = cv2.cvtColor(flipped, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
            image,
            hand_landmarks,
            mp_hands.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(),
            mp_drawing_styles.get_default_hand_connections_style())
      #sending the coordinates of each hand 
      for i in range(len(results.multi_handedness)): 
        if results.multi_handedness[i].classification[0].label == "Right":
          chr=get_sign_right(results,i)
        if results.multi_handedness[i].classification[0].label == "Left":
          chg=get_sign_left(results,i)


    font = cv2.FONT_HERSHEY_SIMPLEX 
    cv2.putText(image,  #writing the value of the right hand
                chr,  
                (400, 50),  
                font, 1,  
                (0, 255, 255),  
                2,  
                cv2.LINE_4) 
    cv2.putText(image,  #writing the value of the left hand
                chg,  
                (50, 50),  
                font, 1,  
                (0, 255, 255),  
                2,  
                cv2.LINE_4) 
    
    cv2.imshow('MediaPipe Hands', image)
    if (cv2.waitKey(5) & 0xFF == 27 ):
      break
cap.release()
cv2.destroyAllWindows()
```
